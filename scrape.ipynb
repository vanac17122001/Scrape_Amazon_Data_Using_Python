# Scrap information of item in Amazon

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Get product name function
def get_productName(soup):
    try:
        product_name = soup.find("span", attrs = {"id":"productTitle"}).text.strip()
    except AttributeError:
        product_name=""
    return product_name
```

```python
# Get product price function
def get_productPrice(soup):
    try:
        product_price = soup.find("span", attrs = {"class":"a-offscreen"}).text.strip()
    except AttributeError:
        product_price=""
    return product_price
```

```python
# Get number rating of product
def get_productNumberRating(soup):
    try:
        product_rating = soup.find("span", attrs = {"id":"acrCustomerReviewText"}).text.strip()
    except AttributeError:
        product_rating=""
    return product_rating
```

```python
# Get rating of product
def get_productStar(soup):
    try:
        product_star = soup.find("span", attrs = {"class":"a-icon-alt"}).text.strip()
    except AttributeError:
        product_star=""
    return product_star
```

```python
# Get rating of product
def get_productAvailability(soup):
    try:
        product_availability = soup.find("div", attrs = {"id":"availability"}).find("span", attrs = {"a-size-medium a-color-success"}).text.strip()
    except AttributeError:
        product_availability=""
    return product_availability
```

```python
if __name__ == '__main__':
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36', 'Accept-Language':'en-US, en;q=0.5'})
    URL='https://www.amazon.com/s?k=playstation+5&sprefix=plays%2Caps%2C338&ref=nb_sb_ss_ts-doa-p_1_5'
    webpage = requests.get(URL, headers=HEADERS)
    # Soup Object containiang all data to html type
    soup = BeautifulSoup(webpage.content, "html.parser")
    # Fetch links as List of Tag Objects
    links = soup.find_all("a", attrs = {"class":"a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})
    
    # create dict to store information of products
    d = {"product_name":[], "price":[], "star":[], "number_rating":[], "availability":[]}

    for i in range(0,len(links),1):
        link = links[i].get('href')
        URL_link = 'https://www.amazon.com' + link
        webpage = requests.get(URL_link, headers=HEADERS)
        soup = BeautifulSoup(webpage.content, "html.parser")
        d['product_name'].append(get_productName(soup))
        d['price'].append(get_productPrice(soup))
        d['star'].append(get_productStar(soup))
        d['number_rating'].append(get_productNumberRating(soup))
        d['availability'].append(get_productAvailability(soup))
    infromationOfProduct_df = pd.DataFrame.from_dict(d)
```

```python
infromationOfProduct_df.head()
```

```python
infromationOfProduct_df['product_name'].replace('', np.nan, inplace=True)
infromationOfProduct_df = infromationOfProduct_df.dropna(subset=['product_name'])
infromationOfProduct_df.to_csv("amazon_informationOfProduct.csv", header=True, index=False)
```
